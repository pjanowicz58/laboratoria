DANE:
  macierz A (n × n)
  wektor b (n)

ELIMINACJA W PRZÓD

for k = 0 .. n-1:

    # wybór pivota
    pivot = k
    for i = k+1 .. n-1:
        jeśli |A[i,k]| > |A[pivot,k]|:
            pivot = i

    # zamiana wierszy
    zamień wiersze A[k] i A[pivot]
    zamień elementy b[k] i b[pivot]

    # zerowanie elementów pod przekątną
    for i = k+1 .. n-1:
        m = A[i,k] / A[k,k]
        for j = k .. n-1:
            A[i,j] = A[i,j] - m * A[k,j]
        b[i] = b[i] - m * b[k]

PODSTAWIENIE WSTECZ

utwórz wektor x długości n

for i = n-1 .. 0:
    s = 0
    for j = i+1 .. n-1:
        s = s + A[i,j] * x[j]
    x[i] = (b[i] - s) / A[i,i]

WYNIK:
  wektor x


```python
def dipole_rho(N: int, sigma: float, amp: float = 1.0):
    yy, xx = np.mgrid[0:N, 0:N]

    x_plus, y_plus = 0.35 * (N - 1), 0.50 * (N - 1)
    x_minus, y_minus = 0.65 * (N - 1), 0.50 * (N - 1)

    rho = amp * np.exp(-((xx - x_plus) ** 2 + (yy - y_plus) ** 2) / (2 * sigma**2))
    rho -= amp * np.exp(-((xx - x_minus) ** 2 + (yy - y_minus) ** 2) / (2 * sigma**2))
    return rho
```
