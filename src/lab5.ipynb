```python
import numpy as np
import matplotlib.pyplot as plt
import csv
```

```python
def save_demo_csv(path: str):
    E = np.arange(1, 251, 2.5)
    S = 3.0 + 0.0025 * E + 18.0 / (E + 1.0) + 0.6 * np.log(1 + 0.02 * E)
    with open(path, 'w', newline='', encoding='utf-8') as f:
        w = csv.writer(f)
        w.writerow(['E_MeV','S_MeV_per_cm'])
        for e, s in zip(E, S):
            w.writerow([e, s])

def load_csv(path: str):
    data = np.genfromtxt(path, delimiter=',', names=True)
    return data['E_MeV'], data['S_MeV_per_cm']

path = 'demo_S.csv'
save_demo_csv(path)
E_data, S_data = load_csv(path)
Emin, E0 = 1.0, 150.0
```
